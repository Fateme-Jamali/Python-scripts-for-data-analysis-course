```python
#problem1
def Problem1(x):
    import numpy as np
    import pandas as pd 
    import matplotlib.pyplot as plt
    %matplotlib inline
    from pandas import Series, DataFrame, MultiIndex
    order_items=pd.read_csv(x + '/' + "olist_order_items_dataset.csv")
    orders=pd.read_csv(x + '/' + "olist_orders_dataset.csv")
    customers=pd.read_csv(x + '/' + "olist_customers_dataset.csv")
    sellers=pd.read_csv(x + '/' + "datasets_55151_195341_olist_sellers_dataset.csv")
    product_translation=pd.read_csv(x + '/' + "datasets_55151_195341_product_category_name_translation.csv")
    products=pd.read_csv(x + '/' + "olist_products_dataset.csv")
    data=pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(order_items,products),product_translation),orders),sellers),customers)
    a=data.groupby('seller_city')['order_id'].count().sort_values( ascending=False)
    aa=a.to_frame()
    #a=aa.rename(columns={'order_id': 'order_count_seller'}, index={'ONE': 'one'})
    b=data.groupby('customer_city')['order_id'].count().sort_values( ascending=False)
    bb=b.to_frame()
    #b=bb.rename(columns={'order_id': 'order_count_customer'}, index={'ONE': 'one'})
    df=aa.head(10)
    dff=bb.head(10)
    #display(df)
    #display(dff)
    result=pd.concat({
    'customer_city': dff.set_index(dff.index), 'seller_city': df.set_index(df.index)
    }, axis=1)
    result.columns = result.columns.droplevel(1)
    #display(result)
    result.plot.bar()
    plt.ylabel("Total Number of orders")
```
